```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import pandas as pd
import mplcursors

import seaborn as sb
sb.set_context("talk")
```

```python
planets = pd.read_csv("TOI_p_greater_200.csv", comment="#")
```

```python
planets
```

```python
plt.figure(figsize = (12,6))
color = "rainbow"

p = plt.scatter(planets["pl_orbper"], planets["pl_rade"], c = planets["pl_eqt"], cmap = color)
plt.xlabel("Orbital Period (Days)")
plt.ylabel("Radii (Earth Radii)")

norm = plt.Normalize(planets["pl_eqt"].min(), planets["pl_eqt"].max())

sm = plt.cm.ScalarMappable(cmap=color, norm=norm)
sm.set_array([])
p.figure.colorbar(sm)
plt.show()
```
